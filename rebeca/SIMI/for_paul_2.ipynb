```python
###replace directories for SIMI_SACT_TUMOUR for synthetic data###

import pandas as pd
import numpy as np

##import SIM I data and create subsets with random selection
SIMI_SACT_TUMOUR=pd.read_table("simulacrum_release_v1/sim_sact_tumour.csv", sep=",")
SIMI_SACT_TUMOUR_50 = SIMI_SACT_TUMOUR.sample(frac =.50)
SIMI_SACT_TUMOUR_25 = SIMI_SACT_TUMOUR.sample(frac =.25)
SIMI_SACT_TUMOUR_10 = SIMI_SACT_TUMOUR.sample(frac =.10)

##new table with counts of unique data
SIMI_SACT_TUMOUR_unique =pd.DataFrame({'Item': ['PRIMARY_DIAGNOSIS', 'CONSULTANT_SPECIALITY_CODE','MORPHOLOGY_CLEAN'], 
                      'Syhtnetic_data_unique_count': [SIMI_SACT_TUMOUR['PRIMARY_DIAGNOSIS'].nunique(), SIMI_SACT_TUMOUR['CONSULTANT_SPECIALITY_CODE'].nunique(),SIMI_SACT_TUMOUR['MORPHOLOGY_CLEAN'].nunique()],
                      '0.50_Syhtnetic_data_unique_count': [SIMI_SACT_TUMOUR_50['PRIMARY_DIAGNOSIS'].nunique(), SIMI_SACT_TUMOUR_50['CONSULTANT_SPECIALITY_CODE'].nunique(), SIMI_SACT_TUMOUR_50['MORPHOLOGY_CLEAN'].nunique()],
                     '0.25_Syhtnetic_data_unique_count': [SIMI_SACT_TUMOUR_25['PRIMARY_DIAGNOSIS'].nunique(), SIMI_SACT_TUMOUR_25['CONSULTANT_SPECIALITY_CODE'].nunique(), SIMI_SACT_TUMOUR_25['MORPHOLOGY_CLEAN'].nunique()],
                    '0.10_Syhtnetic_data_unique_count': [SIMI_SACT_TUMOUR_10['PRIMARY_DIAGNOSIS'].nunique(), SIMI_SACT_TUMOUR_10['CONSULTANT_SPECIALITY_CODE'].nunique(), SIMI_SACT_TUMOUR_10['MORPHOLOGY_CLEAN'].nunique()]
                                      })
SIMI_SACT_TUMOUR_unique['Missing_values_0.50_count']=SIMI_SACT_TUMOUR_unique['Syhtnetic_data_unique_count']-SIMI_SACT_TUMOUR_unique['0.50_Syhtnetic_data_unique_count']
SIMI_SACT_TUMOUR_unique['Missing_values_0.50_percentage']=(SIMI_SACT_TUMOUR_unique['Syhtnetic_data_unique_count']-SIMI_SACT_TUMOUR_unique['0.50_Syhtnetic_data_unique_count'])/SIMI_SACT_TUMOUR_unique['Syhtnetic_data_unique_count']*100
SIMI_SACT_TUMOUR_unique['Missing_values_0.25_count']=SIMI_SACT_TUMOUR_unique['Syhtnetic_data_unique_count']-SIMI_SACT_TUMOUR_unique['0.25_Syhtnetic_data_unique_count']
SIMI_SACT_TUMOUR_unique['Missing_values_0.25_percentage']=(SIMI_SACT_TUMOUR_unique['Syhtnetic_data_unique_count']-SIMI_SACT_TUMOUR_unique['0.25_Syhtnetic_data_unique_count'])/SIMI_SACT_TUMOUR_unique['Syhtnetic_data_unique_count']*100
SIMI_SACT_TUMOUR_unique['Missing_values_0.10_count']=SIMI_SACT_TUMOUR_unique['Syhtnetic_data_unique_count']-SIMI_SACT_TUMOUR_unique['0.10_Syhtnetic_data_unique_count']
SIMI_SACT_TUMOUR_unique['Missing_values_0.10_percentage']=(SIMI_SACT_TUMOUR_unique['Syhtnetic_data_unique_count']-SIMI_SACT_TUMOUR_unique['0.10_Syhtnetic_data_unique_count'])/SIMI_SACT_TUMOUR_unique['Syhtnetic_data_unique_count']*100

print(SIMI_SACT_TUMOUR_unique)

##make frequency tables for SIM I and subsets
def frequency (table, column)
table_column_CODE_FREQ=pd.value_counts(table[column]).to_frame().reset_index()
table_column_CODE_FREQ.columns=[column,'count_synth']

table_50_column_CODE_FREQ=pd.value_counts(table[column]).to_frame().reset_index()
table_50_column_CODE_FREQ.columns=[column,'count_0.50_synth']

table_25_column_CODE_FREQ=pd.value_counts(table[column]).to_frame().reset_index()
table_25_column_CODE_FREQ.columns=[column,'count_0.25_synth']

table_10_column_CODE_FREQ=pd.value_counts(table[column]).to_frame().reset_index()
table_10_column_CODE_FREQ.columns=[column,'count_0.10_synth']

return table_column_CODE_FREQ, table_50_column_CODE_FREQ, table_25_column_CODE_FREQ,table_10_column_CODE_FREQ

frequency()


SIMI_SACT_TUMOUR_CONSULTANT_SPECIALITY_CODE_FREQ=pd.value_counts(SIMI_SACT_TUMOUR['CONSULTANT_SPECIALITY_CODE']).to_frame().reset_index()
SIMI_SACT_TUMOUR_CONSULTANT_SPECIALITY_CODE_FREQ.columns=['CONSULTANT_SPECIALITY_CODE','count_synth']

SIMI_SACT_TUMOUR_50_CONSULTANT_SPECIALITY_CODE_FREQ=pd.value_counts(SIMI_SACT_TUMOUR_50['CONSULTANT_SPECIALITY_CODE']).to_frame().reset_index()
SIMI_SACT_TUMOUR_50_CONSULTANT_SPECIALITY_CODE_FREQ.columns=['CONSULTANT_SPECIALITY_CODE','count_0.50_synth']

SIMI_SACT_TUMOUR_25_CONSULTANT_SPECIALITY_CODE_FREQ=pd.value_counts(SIMI_SACT_TUMOUR_25['CONSULTANT_SPECIALITY_CODE']).to_frame().reset_index()
SIMI_SACT_TUMOUR_25_CONSULTANT_SPECIALITY_CODE_FREQ.columns=['CONSULTANT_SPECIALITY_CODE','count_0.25_synth']

SIMI_SACT_TUMOUR_10_CONSULTANT_SPECIALITY_CODE_FREQ=pd.value_counts(SIMI_SACT_TUMOUR_10['CONSULTANT_SPECIALITY_CODE']).to_frame().reset_index()
SIMI_SACT_TUMOUR_10_CONSULTANT_SPECIALITY_CODE_FREQ.columns=['CONSULTANT_SPECIALITY_CODE','count_0.10_synth']

#

SIMI_SACT_TUMOUR_PRIMARY_DIAGNOSIS_FREQ=pd.value_counts(SIMI_SACT_TUMOUR['PRIMARY_DIAGNOSIS']).to_frame().reset_index()
SIMI_SACT_TUMOUR_PRIMARY_DIAGNOSIS_FREQ.columns=['PRIMARY_DIAGNOSIS','count_synth']

SIMI_SACT_TUMOUR_50_PRIMARY_DIAGNOSIS_FREQ=pd.value_counts(SIMI_SACT_TUMOUR_50['PRIMARY_DIAGNOSIS']).to_frame().reset_index()
SIMI_SACT_TUMOUR_50_PRIMARY_DIAGNOSIS_FREQ.columns=['PRIMARY_DIAGNOSIS','count_0.50_synth']

SIMI_SACT_TUMOUR_25_PRIMARY_DIAGNOSIS_FREQ=pd.value_counts(SIMI_SACT_TUMOUR_25['PRIMARY_DIAGNOSIS']).to_frame().reset_index()
SIMI_SACT_TUMOUR_25_PRIMARY_DIAGNOSIS_FREQ.columns=['PRIMARY_DIAGNOSIS','count_0.25_synth']

SIMI_SACT_TUMOUR_10_PRIMARY_DIAGNOSIS_FREQ=pd.value_counts(SIMI_SACT_TUMOUR_10['PRIMARY_DIAGNOSIS']).to_frame().reset_index()
SIMI_SACT_TUMOUR_10_PRIMARY_DIAGNOSIS_FREQ.columns=['PRIMARY_DIAGNOSIS','count_0.10_synth']

#

SIMI_SACT_TUMOUR_MORPHOLOGY_CLEAN_FREQ=pd.value_counts(SIMI_SACT_TUMOUR['MORPHOLOGY_CLEAN']).to_frame().reset_index()
SIMI_SACT_TUMOUR_MORPHOLOGY_CLEAN_FREQ.columns=['MORPHOLOGY_CLEAN','count_synth']

SIMI_SACT_TUMOUR_50_MORPHOLOGY_CLEAN_FREQ=pd.value_counts(SIMI_SACT_TUMOUR_50['MORPHOLOGY_CLEAN']).to_frame().reset_index()
SIMI_SACT_TUMOUR_50_MORPHOLOGY_CLEAN_FREQ.columns=['MORPHOLOGY_CLEAN','count_0.50_synth']

SIMI_SACT_TUMOUR_25_MORPHOLOGY_CLEAN_FREQ=pd.value_counts(SIMI_SACT_TUMOUR_25['MORPHOLOGY_CLEAN']).to_frame().reset_index()
SIMI_SACT_TUMOUR_25_MORPHOLOGY_CLEAN_FREQ.columns=['MORPHOLOGY_CLEAN','count_0.25_synth']

SIMI_SACT_TUMOUR_10_MORPHOLOGY_CLEAN_FREQ=pd.value_counts(SIMI_SACT_TUMOUR_10['MORPHOLOGY_CLEAN']).to_frame().reset_index()
SIMI_SACT_TUMOUR_10_MORPHOLOGY_CLEAN_FREQ.columns=['MORPHOLOGY_CLEAN','count_0.10_synth']

##merging the frequency tables
CONSULTANT_SPECIALITY_CODE_FREQ_merge=pd.merge(SIMI_SACT_TUMOUR_CONSULTANT_SPECIALITY_CODE_FREQ, SIMI_SACT_TUMOUR_50_CONSULTANT_SPECIALITY_CODE_FREQ, on='CONSULTANT_SPECIALITY_CODE',how='outer')
CONSULTANT_SPECIALITY_CODE_FREQ_merge=pd.merge(CONSULTANT_SPECIALITY_CODE_FREQ_merge, SIMI_SACT_TUMOUR_25_CONSULTANT_SPECIALITY_CODE_FREQ, on='CONSULTANT_SPECIALITY_CODE',how='outer')
CONSULTANT_SPECIALITY_CODE_FREQ_merge=pd.merge(CONSULTANT_SPECIALITY_CODE_FREQ_merge, SIMI_SACT_TUMOUR_10_CONSULTANT_SPECIALITY_CODE_FREQ, on='CONSULTANT_SPECIALITY_CODE',how='outer')
CONSULTANT_SPECIALITY_CODE_FREQ_merge.sort_values(by=['count_synth'],ascending=False)

PRIMARY_DIAGNOSIS_FREQ_merge=pd.merge(SIMI_SACT_TUMOUR_PRIMARY_DIAGNOSIS_FREQ, SIMI_SACT_TUMOUR_50_PRIMARY_DIAGNOSIS_FREQ, on='PRIMARY_DIAGNOSIS',how='outer')
PRIMARY_DIAGNOSIS_FREQ_merge=pd.merge(PRIMARY_DIAGNOSIS_FREQ_merge, SIMI_SACT_TUMOUR_25_PRIMARY_DIAGNOSIS_FREQ, on='PRIMARY_DIAGNOSIS',how='outer')
PRIMARY_DIAGNOSIS_FREQ_merge=pd.merge(PRIMARY_DIAGNOSIS_FREQ_merge, SIMI_SACT_TUMOUR_10_PRIMARY_DIAGNOSIS_FREQ, on='PRIMARY_DIAGNOSIS',how='outer')
PRIMARY_DIAGNOSIS_FREQ_merge.sort_values(by=['count_synth'],ascending=False)

MORPHOLOGY_CLEAN_FREQ_merge=pd.merge(SIMI_SACT_TUMOUR_MORPHOLOGY_CLEAN_FREQ, SIMI_SACT_TUMOUR_50_MORPHOLOGY_CLEAN_FREQ, on='MORPHOLOGY_CLEAN',how='outer')
MORPHOLOGY_CLEAN_FREQ_merge=pd.merge(MORPHOLOGY_CLEAN_FREQ_merge, SIMI_SACT_TUMOUR_25_MORPHOLOGY_CLEAN_FREQ, on='MORPHOLOGY_CLEAN',how='outer')
MORPHOLOGY_CLEAN_FREQ_merge=pd.merge(MORPHOLOGY_CLEAN_FREQ_merge, SIMI_SACT_TUMOUR_10_MORPHOLOGY_CLEAN_FREQ, on='MORPHOLOGY_CLEAN',how='outer')
MORPHOLOGY_CLEAN_FREQ_merge.sort_values(by=['count_synth'],ascending=False)

```
